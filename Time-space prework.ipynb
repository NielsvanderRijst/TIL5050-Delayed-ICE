```python
# Make a long DataFrame of the planned timetable of all trains between the ICE you put in below and the next one.
# If ICE_value = 140 (the last train of the day), only the 2 hours after ICE 140 is plotted.
ICE_value = 240

# Load the DataFrame with the corresponding trains
df = pd.read_csv(f"timetables/timetable_{ICE_value}.csv", sep=";", engine="python")

# Row 0 contains the colors
colors = df.iloc[0].copy()

# The rest of the rows contain information about stations and stopping times
df2 = df.iloc[1:].copy()

# Since the colors are now placed into a new DataFrame, this cell needs to be 'Station'
df2.rename(columns={"train_id": "station"}, inplace=True)

# Build the new DataFrame
records = []

for col in df2.columns:
    if col == "station" or col.startswith("Unnamed"):
        continue

    # The color of the column needs to be obtained
    color = colors[col]

    # Next, loop through all the rows to obtain the stations and times ('07:41')
    for _, row in df2.iterrows():
        st = row["station"]
        time = row[col]

        # Only if there is a time (not NaN), a new row of the DataFrame is added with the train_id, station, time and color
        if isinstance(time, str):
            records.append({
                "train_id": col,
                "station": st,
                "time": time,
                "color": color
            })

# Saving the DataFrame
planned_timetable = pd.DataFrame(records)
planned_timetable.to_csv(f"Planned_timetables/planned_timetable_{ICE_value}.csv", index=False)
```
